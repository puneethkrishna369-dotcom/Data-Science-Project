#####  Display 5 records where launch sites begin with the string 'CCA'

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# Filter for NASA (CRS) launches
nasa_crs = df[df['Customer'] == 'NASA (CRS)']

# Calculate total payload mass
total_payload = nasa_crs['PAYLOAD_MASS__KG_'].sum()
print("Total payload mass carried by NASA (CRS) boosters:", total_payload, "kg")
```

```python
%%sql SELECT SUM(PAYLOAD_MASS__KG_) AS TotalPayloadMass
FROM SPACEXTBL
WHERE Customer = 'NASA (CRS)';
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# Filter for booster version F9 v1.1
f9_v11 = df[df['Booster_Version'] == 'F9 v1.1']

# Calculate average payload mass
avg_payload = f9_v11['PAYLOAD_MASS__KG_'].mean()
print("Average payload mass carried by F9 v1.1 boosters:", avg_payload, "kg")
```

```python
%%sql SELECT AVG(PAYLOAD_MASS__KG_) AS AvgPayloadMass
FROM SPACEXTBL
WHERE Booster_Version = 'F9 v1.1';
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
# Filter for successful ground pad landings
ground_pad_success = df[df['Landing_Outcome'] == 'Success (ground pad)']

# Find the earliest date
first_success_date = ground_pad_success['Date'].min()
print("First successful ground pad landing date:", first_success_date)
```

```python
%%sql SELECT MIN(Date) AS FirstSuccessDate
FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (ground pad)';
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# Filter boosters with success in drone ship and payload mass between 4000 and 6000
filtered_boosters = df[
    (df['Landing_Outcome'] == 'Success (drone ship)') &
    (df['PAYLOAD_MASS__KG_'] > 4000) &
    (df['PAYLOAD_MASS__KG_'] < 6000)
]

# Display the booster names
print(filtered_boosters['Booster_Version'].unique())
```

```python
%%sql SELECT DISTINCT Booster_Version
FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (drone ship)'
  AND PAYLOAD_MASS__KG_ > 4000
  AND PAYLOAD_MASS__KG_ < 6000;
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# Count mission outcomes
outcome_counts = df['Mission_Outcome'].value_counts()
print(outcome_counts)
```

```python
%%sql SELECT Mission_Outcome, COUNT(*) AS OutcomeCount
FROM SPACEXTBL
GROUP BY Mission_Outcome
ORDER BY OutcomeCount DESC;
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
# Find the maximum payload mass
max_payload = df['PAYLOAD_MASS__KG_'].max()

# Get booster versions that carried this payload
boosters = df[df['PAYLOAD_MASS__KG_'] == max_payload]['Booster_Version'].unique()
print("Boosters with maximum payload mass:", boosters)
```

```python
%%sql SELECT DISTINCT Booster_Version
FROM SPACEXTBL
WHERE PAYLOAD_MASS__KG_ = (
    SELECT MAX(PAYLOAD_MASS__KG_)
    FROM SPACEXTBL
);
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
import pandas as pd

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter for year 2015 and failure in drone ship
filtered = df[
    (df['Date'].dt.year == 2015) &
    (df['Landing_Outcome'].str.contains('Failure (drone ship)', na=False, regex=False))
]

# Add month names
filtered['MonthName'] = filtered['Date'].dt.month_name()

# Select relevant columns
result = filtered[['MonthName', 'Booster_Version', 'Launch_Site', 'Landing_Outcome']]

print(result)
```

```python
%%sql SELECT 
    CASE substr(Date, 6, 2)
        WHEN '01' THEN 'January'
        WHEN '02' THEN 'February'
        WHEN '03' THEN 'March'
        WHEN '04' THEN 'April'
        WHEN '05' THEN 'May'
        WHEN '06' THEN 'June'
        WHEN '07' THEN 'July'
        WHEN '08' THEN 'August'
        WHEN '09' THEN 'September'
        WHEN '10' THEN 'October'
        WHEN '11' THEN 'November'
        WHEN '12' THEN 'December'
    END AS MonthName,
    Booster_Version,
    Launch_Site,
    Landing_Outcome
FROM SPACEXTBL
WHERE substr(Date, 1, 4) = '2015'
  AND Landing_Outcome LIKE '%Failure%'
  AND Landing_Outcome LIKE '%drone ship%';
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
import pandas as pd

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter rows between the given date range
mask = (df['Date'] >= '2010-06-04') & (df['Date'] <= '2017-03-20')
filtered = df[mask]

# Count landing outcomes and sort in descending order
outcome_counts = filtered['Landing_Outcome'].value_counts().sort_values(ascending=False)

print(outcome_counts)
```

```python
%%sql SELECT Landing_Outcome, COUNT(*) AS OutcomeCount
FROM SPACEXTBL
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY OutcomeCount DESC;
```
